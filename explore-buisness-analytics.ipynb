```python
# Importing libraries

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('bmh')
```

```python
df = pd.read_csv("SampleSuperstore.csv")
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
df.describe()
```

```python
sns.jointplot(x = 'Sales', y = 'Quantity', data = df)
```

```python
df.isnull().any()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(keep = 'first', inplace = True)
df
```

```python
sns.countplot(x = 'Profit', data = df)
```

```python
corr = df.corr()
corr
```

```python
sns.heatmap(corr, annot = True)
```

```python
sns.countplot(df['Ship Mode'])
plt.show()
```

```python
sns.countplot(df['Segment'])
plt.show()
```

```python
df.hist(figsize = (10,10), bins = 50)
plt.show()
```

```python
plt.figure(figsize = (12,17))
plt.scatter(x = df["Sales"], y = df["Profit"], c = "m")
plt.title("Sales Vs Profit")
plt.xlabel("Total Sale")
plt.ylabel("Profit")
plt.show()
```

```python
plt.figure(figsize = (10,8))
df["Sub-Category"].value_counts().plot.pie(autopct = '%1.1f%%')
plt.show()
```

```python
plt.figure(figsize = (11,7))
df.groupby('Sub-Category')['Profit', 'Sales'].agg(['sum']).plot.bar()
plt.show()
```

```python
plt.figure(figsize = (12,7))
sns.countplot(x = 'Sub-Category', hue = 'Region', data = df)
plt.xticks(rotation = 90)
plt.show()
```

Here we can easily conclude how much effect our data set variable in a buisness with various graphs and we can compare overall statistics our data in terms of retailer buisness analytics
